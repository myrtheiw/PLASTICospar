```python
#load the data path
data = r"...\dataset_semifinal\data.yaml"
```

## MODEL TRAINING

```python
# Load a model
model = YOLO('yolov8l.yaml')  # build a new model from YAML


# train the model
results = model.train(
data=data,
project= "plastic_semifinal",
optimizer = "Adam",
batch= 2,
imgsz = 1280,
save_period=1,
save_json=True,
epochs=300
)
```
